```python
import requests
import pandas as pd
```

```python
credito = pd.read_csv("Credit.csv")
credito
```

```python
for col in credito.columns:
    if credito[col].dtype == 'object':
        credito[col] = credito[col].astype('category').cat.codes
```

```python
dados = credito.iloc[0:10, :20].to_json(orient='records')
dados
```

```python
#EXECUTE mlflow models serve --model-uri runs:/{model id}/{experiment name} -p {port number} --env-manager=local
pred = requests.post(url='http://localhost:2345/invocations', headers={'Content-Type': 'application/json'}, data=dados)
```
